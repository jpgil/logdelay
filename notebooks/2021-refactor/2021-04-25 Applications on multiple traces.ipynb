Let be $L=\{ T_i \ | i=1 ... N\}$ a set of $N$ traces in the alphabet $\Sigma$.

*Intuition*: 

$$
\forall e=ab \in \Sigma^2 \\
\eta_L(e) = \sum_{ T_i \in L } { \eta_{T_i}(e) } 
$$

```python
T1 = 'abc'
Gp1 = pair_graph(T1)
print(Gp1.pd)
print('Sequences: {}'.format( sequences_from_pair_graph( Gp1 )))
graph(Gp1)
```

## Some examples

```python
def sequences_from_log( L ):
    Gp = pair_graph('')
    for Ti in L:
        Gp = Gp + pair_graph(Ti)
    return sequences_from_pair_graph( Gp )
```

```python
Pg = pair_graph('')
sequences_from_pair_graph( Pg )
```

```python
T1 = 'abababababab'; T2 = 'ba'
sequences_from_pair_graph( pair_graph(T1) + pair_graph(T2) )
```

```python
L=['abc', 'XY', 'abXYcaXbcY']
sequences_from_log(L)
```

```python
L=['abcde', 'abcXY', '123abc']
sequences_from_log(L)
```
